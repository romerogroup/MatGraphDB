```python
from poly_graphs_lib.utils import PROJECT_DIR
import os
import json
import numpy as np
from glob import glob 

import pymatgen.core as pmat

from poly_graphs_lib.database.json import DB_DIR

database_files=glob(DB_DIR + '\*.json')
```

```python
database_files=glob(DB_DIR + '\*.json')
```

```python
success_files=[]
failed_files=[]
for file in database_files:
    with open(file, 'r') as f:
        data = json.load(f)

        if data["coordination_environments_multi_weight"] is not None:
            success_files.append(file)
        else:
            failed_files.append(file)

print("Success: ", len(success_files))
print("Failed: ", len(failed_files))

print("Success: ", success_files[:10])
print("Failed: ", failed_files[:10])
```

```python
from pymatgen.analysis.chemenv.coordination_environments.chemenv_strategies import SimplestChemenvStrategy,MultiWeightsChemenvStrategy
from pymatgen.analysis.chemenv.coordination_environments.coordination_geometry_finder import LocalGeometryFinder
from pymatgen.analysis.chemenv.coordination_environments.structure_environments import LightStructureEnvironments
import copy

file=failed_files[1]
with open(file) as f:
    db = json.load(f)
    struct = pmat.Structure.from_dict(db['structure'])


print(struct.atomic_numbers)
print(struct.cart_coords)
lgf = LocalGeometryFinder()
lgf.setup_structure(structure=struct)


se = lgf.compute_structure_environments(maximum_distance_factor=1.41,
                                        minimum_angle_factor=0.05,
                                        voronoi_normalized_distance_tolerance=0.05,
                                        voronoi_normalized_angle_tolerance=0.03,
                                        only_cations=False)
strategy = SimplestChemenvStrategy(distance_cutoff=1.41, angle_cutoff=0.3)
# se = lgf.compute_structure_environments(maximum_distance_factor=2.5,only_cations=False)
# strategy =SimpleAbundanceChemenvStrategy()
# strategy = MultiWeightsChemenvStrategy.stats_article_weights_parameters()
lse = LightStructureEnvironments.from_structure_environments(strategy=strategy, structure_environments=se)



coordination_environments = copy.copy(lse.coordination_environments)
for i,env in enumerate(lse.coordination_environments):
    if env is None:
        print(env)
        coordination_environments[i]=[{'ce_symbol': 'S:1', 'ce_fraction': 1.0, 'csm': 0.0, 'permutation': [0]}]
        # env=[{'ce_symbol': 'UKNOWN:1', 'ce_fraction': 0.999999999677424, 'csm': 0.0, 'permutation': []}]

print(coordination_environments)
```
