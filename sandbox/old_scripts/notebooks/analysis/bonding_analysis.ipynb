```python
import os
import json
from glob import glob

import numpy as np

import pymatgen.core as pmat

from poly_graphs_lib.utils import PROJECT_DIR
```

```python
database_dir=os.path.join(PROJECT_DIR,'data','raw','mp_database')
database_files=glob(database_dir + '\*.json')
file=database_files[0]
file=os.path.join(database_dir,'mp-170.json')
print(file.split(os.sep)[-1].split('.')[0])
with open(file) as f:
    db = json.load(f)
    struct = pmat.Structure.from_dict(db['structure'])
```

```python
from pymatgen.analysis.local_env import CovalentBondNN, CrystalNN, MinimumDistanceNN,CutOffDictNN
from poly_graphs_lib.utils.periodic_table import covalent_cutoff_map
```

```python
cutoff_dict=covalent_cutoff_map(tol=0)
print(list(cutoff_dict.keys()))
print(cutoff_dict[('Ho','Ho')])
```

```python
cutoff_nn=CutOffDictNN(cut_off_dict=cutoff_dict)
all_nn=cutoff_nn.get_all_nn_info(structure=struct)
all_nn
print(len(all_nn))
```

```python
nearest_neighbors=[]
for site_nn in all_nn:
    print('_____________________________')
    neighbor_index=[]
    for nn in site_nn:

        index=nn['site_index']
        neighbor_index.append(index)
    nearest_neighbors.append(neighbor_index)
print(nearest_neighbors)
```
