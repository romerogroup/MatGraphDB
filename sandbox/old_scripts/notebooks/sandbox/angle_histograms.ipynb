### Plotting different sigma values

```python
n_bins =100
min_val=np.pi/8
max_val=(3/2)*np.pi

sigmas = [0.5,0.1,0.01,0.05]

fig,axs = plt.subplots(len(sigmas))
for i,sigma in enumerate(sigmas):

    poly = PolyFeaturizer(vertices=test_polys[0])
    x = np.linspace(min_val,max_val,n_bins)
    vert_histograms = poly.get_verts_neighbor_angles_encodings(n_bins = n_bins, min_val=min_val, max_val=max_val, sigma=sigma)

    for vert_histogram in vert_histograms:
        axs[i].plot(x,vert_histogram)

    axs[i].set_title(f'Sigma = {sigma}')
    if i != len(sigmas) -1:
        axs[i].get_xaxis().set_visible(False)

plt.tight_layout()
```

### Plotting normalization scheme

```python
n_bins =100
min_val=0
max_val=(3/2)*np.pi
sigma=0.1


normals = ['None','normal','integral']

i_poly = [0,1,2,3]
polys = [test_polys[i] for i in i_poly]
poly_names = [test_names[i] for i in i_poly]
fig,axs = plt.subplots(len(normals))

for i, normal in enumerate(normals):
    for j,(poly,poly_name) in enumerate(zip(polys,poly_names)):

        poly = PolyFeaturizer(vertices=poly)
        x = np.linspace(min_val,max_val,n_bins)
        vert_histograms = poly.get_verts_neighbor_angles_encodings(n_bins = n_bins, min_val=min_val, max_val=max_val, sigma=sigma)
        
        for vert_histogram in vert_histograms:
            if normal == 'integral':
                integral = np.trapz(vert_histogram,x)
                vert_histogram=vert_histogram/integral
            elif normal == 'normal':
                vert_histogram=vert_histogram/max(vert_histogram)
            elif normal == 'None':
                vert_histogram=vert_histogram
            axs[i].plot(x,vert_histogram)

        axs[i].set_title(f'Normalization : {normal}')
        if i != len(normals) -1:
            axs[i].get_xaxis().set_visible(False)

fig.suptitle("Normalization scheme")
plt.tight_layout()
```
